After combining everything into the Schedule pacing, it seems that we have values that are not on a number scale, are numbers as string, or are #REF errors (lol). To fix this we must create a scale that can map the strings to numbers, all numbers as strings to numbers, and remove any extraneous information

While subjective, the scale that I think is most fitting is as follows:

* Just right - 5
* A little too slow - 4
* A little too fast - 3
* Way too slow - 2
* Way too fast - 1

As for the Numbers as Strings, we're going to just convert them into numbers

```python
   scaled_values = {
        'Just right': 5,
        'A little too slow': 4,
        'A little too fast': 3,
        'Way too slow': 2,
        'Way too fast': 1
    }
    
def fix_pacing(row):
 
    if row in scaled_values:
        return scaled_values[row]
    else:
        if type(row) == type('a') and row != '#REF!':
            return int(row)
        elif type(row) == None:
            return None
        else:
            return row
            
```

```python
df['Schedule Pacing'].map(fix_pacing).unique()
```

As we can see, using the transformations that we've applied we now get a much cleaner dataset in which we can refine much more easily later. Which leaves us to applying these changes to the dataset.

```python
df['Schedule Pacing'] = df['Schedule Pacing'].map(fix_pacing)
```

```python
target_len = len('summer')
def fix_track(row):
    if type(row) == type('a'):
        if row[:target_len] == 'summer':
            return str(row[target_len:])
        else:
            return row         
```
